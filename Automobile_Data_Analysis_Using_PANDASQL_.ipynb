**Question - 1. What is the average price and average mileage for convertiable cars?**

````python
# Question - 1. What is the average price for convertiable cars?
A = pd.read_sql_query('''

SELECT
    make,
    AVG(price) AS avg_price
FROM
    automobile_data
WHERE
    "body-style" = 'convertible'
    AND price IS NOT NULL
GROUP BY
    make
ORDER BY
    avg_price DESC

''', conn)
A
````

**Question - 2. What is the average price for all type of cars?**

````python
# Question - 2. What is the average price for all type of cars?
B = pd.read_sql_query('''

SELECT
    "body-style",
    AVG(price) AS avg_price
FROM
    automobile_data
GROUP BY
    "body-style"
ORDER BY
    avg_price DESC



''', conn)
B
````

**Question - 3. What is the average mileage for all type of cars?**

````python
# Question - 3. What is the average mileage for all type of cars?
C = pd.read_sql_query('''

SELECT
    "body-style",
    AVG("city-mpg" + "highway-mpg") / 2.0 AS avg_mileage
FROM
    automobile_data
WHERE
    "city-mpg" IS NOT NULL AND "highway-mpg" IS NOT NULL
GROUP BY
    "body-style"
ORDER BY
    avg_mileage DESC

''', conn)
C
````

**Question - 4. Sort car by company and price in decreasing order?**

````python
# Question - 4. Sort car by company and price in decreasing order?
D = pd.read_sql_query('''

SELECT
    make,
    price
FROM
    automobile_data
WHERE
    price IS NOT NULL
ORDER BY
    make ASC, price DESC
limit 20

''', conn)
D
````

**Question - 6. Which company is producing most expensive car?**

````python
# Question - 6. Which company is producing most expensive car?
F = pd.read_sql_query('''

SELECT
    make,
    AVG(price) AS avg_price
FROM
    automobile_data
WHERE
    price IS NOT NULL
GROUP BY
    make
ORDER BY
    avg_price DESC
LIMIT 5

''', conn)
F
````

**Question - 7. Name the cars on the basis of maximum lenght, breadth and height?**

````python
# Question - 7. Name the cars on the basis of maximum length, breadth and height?
# Convert necessary columns to numeric
df['length'] = pd.to_numeric(df['length'], errors='coerce')
df['width'] = pd.to_numeric(df['width'], errors='coerce')
df['height'] = pd.to_numeric(df['height'], errors='coerce')

G = pd.read_sql_query('''

SELECT
    make,
    length,
    width,
    height
FROM
    automobile_data
WHERE
    length IS NOT NULL AND width IS NOT NULL AND height IS NOT NULL
ORDER BY
    length DESC, width DESC, height DESC
LIMIT 10

''', conn)
G
````

**Question - 8. Name of the car for maximum horsepower in descending order?**

````python
# Question - 8. Name of the car for maximum horsepower in descending order?
# Clean horsepower column
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

H = pd.read_sql_query('''

SELECT
    make || ' ' || "car_model" AS car_name,
    horsepower
FROM
    automobile_data
WHERE
    horsepower IS NOT NULL
ORDER BY
    horsepower DESC
limit 10

''', conn)
H
````

**Question - 9. Find the number of car by type of engine?**

````python
# Question - 9. Find the number of car by type of engine?
# Clean engine-type column
df['engine-type'] = df['engine-type'].astype(str).str.strip()

I = pd.read_sql_query('''

SELECT
    "engine-type" AS engine_type,
    COUNT(*) AS car_count
FROM
    automobile_data
WHERE
    "engine-type" IS NOT NULL
GROUP BY
    "engine-type"
ORDER BY
    car_count DESC

''', conn)
I
````

**Question - 10. Which car has which type of fuel-system?**



````python
# Question - 10. Which car has which type of fuel-system?
J = pd.read_sql_query('''

select make, "fuel-system" from automobile_data
group by make
limit 10

''', conn)
J
````

**Question - 11. Find the number of car  by type of fuel system?**

````python
# Question - 11. Find the number of car  by type of fuel system?
# Clean fuel-system column
df['fuel-system'] = df['fuel-system'].astype(str).str.strip()

K = pd.read_sql_query('''

SELECT
    "fuel-system" AS fuel_system,
    COUNT(*) AS car_count
FROM
    automobile_data
WHERE
    "fuel-system" IS NOT NULL
GROUP BY
    "fuel-system"
ORDER BY
    car_count DESC

''', conn)
K
````

**Question - 12. Find normalised losses for each company?**

````python
# Question - 12. Find normalised losses for each company?
# Clean and prepare data
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df['make'] = df['make'].astype(str).str.strip()

L = pd.read_sql_query('''

SELECT
    make AS company,
    AVG("normalized-losses") AS avg_normalized_losses
FROM
    automobile_data
WHERE
    "normalized-losses" IS NOT NULL
GROUP BY
    make
ORDER BY
    avg_normalized_losses DESC

''', conn)
L
````

**Question - 13. Find the car having average peak-rmp??**

````python
# Question - 13. Find the car having average peak-rmp?
# Data cleaning
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'], errors='coerce')
df = df.dropna(subset=['peak-rpm'])  # Remove missing values
df['make'] = df['make'].astype(str).str.strip()


M = pd.read_sql_query('''

SELECT
    make || ' ' || "body-style" AS car,
    AVG("peak-rpm") AS avg_peak_rpm
FROM
    automobile_data
WHERE
    "peak-rpm" IS NOT NULL
GROUP BY
    car
ORDER BY
    avg_peak_rpm DESC
limit 10

''', conn)
M
````

**Question - 14. Company wise Number of cars?**

````python
# Question - 14. Company wise Number of cars?
# Clean company name
df['make'] = df['make'].astype(str).str.strip()

N = pd.read_sql_query('''

SELECT
    make AS company,
    COUNT(*) AS car_count
FROM
    automobile_data
WHERE
    make IS NOT NULL
GROUP BY
    make
ORDER BY
    car_count DESC
LIMIT 15

''', conn)
N
````

**Question - 15. Company wise average horse power?**

````python
# Question - 15. Company wise average horse power?
# Clean and convert horsepower column
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df = df.dropna(subset=['horsepower'])
df['make'] = df['make'].astype(str).str.strip()

O = pd.read_sql_query('''

SELECT
    make AS company,
    AVG(horsepower) AS avg_horsepower
FROM
    automobile_data
WHERE
    horsepower IS NOT NULL
GROUP BY
    make
ORDER BY
    avg_horsepower DESC
LIMIT 15


''', conn)
O
````

**Question - 16. Number of car by body style?**

````python
# Question - 16. Number of car by body style?
# Clean data
df['body-style'] = df['body-style'].astype(str).str.strip()

P = pd.read_sql_query('''

SELECT
    "body-style" AS body_style,
    COUNT(*) AS car_count
FROM
    automobile_data
WHERE
    "body-style" IS NOT NULL
GROUP BY
    "body-style"

''', conn)
P
````

**Question - 17. No. of cylinedrs by car?**

````python
# Question - 17. No. of cylinedrs by car?

# Clean and convert 'num-of-cylinders' to a consistent numeric or categorical form
df['num-of-cylinders'] = df['num-of-cylinders'].astype(str).str.strip()

Q = pd.read_sql_query('''

SELECT
    "num-of-cylinders" AS cylinders,
    COUNT(*) AS car_count
FROM
    automobile_data
WHERE
    "num-of-cylinders" IS NOT NULL
GROUP BY
    "num-of-cylinders"
ORDER BY
    car_count DESC


''', conn)
Q
````

#### Chart - 17 - Pair Plot

````python
# Pair Plot visualization code
# Clean column names (optional but helpful)
df.columns = df.columns.str.strip().str.lower().str.replace('-', '_')

# Convert numeric columns to proper format
numeric_columns = [
    'price', 'engine_size', 'horsepower', 'curb_weight',
    'city_mpg', 'highway_mpg', 'length', 'width', 'height'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows with missing values in selected columns
plot_df = df[numeric_columns].dropna()

# Create pair plot
sns.pairplot(plot_df)
plt.suptitle("Pair Plot of Key Automobile Features", y=1.02)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

**Answer Here.**

**I picked the pair plot because:**

✅ Multiple Relationships in One View

* It shows all pairwise comparisons between selected numeric features, helping identify:

1 .Linear relationships (e.g., engine size vs. price)

2 .Clusters or outliers

3 .Correlated features for model selection

✅ Distribution + Interaction

* You get both: Individual distributions on the diagonal

* Scatter plots off-diagonal showing how features relate

✅ Ideal for Exploratory Data Analysis (EDA)

* Pair plots are a powerful tool in the early stages of analysis to:

* Understand data structure

* Detect redundancies

* Spot trends across multiple features quickly
